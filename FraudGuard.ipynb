```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the banknote authentication dataset
df = pd.read_csv("data_banknote_authentication.txt", header=None, names=['Variance', 'Skewness', 'Curtosis', 'Entropy', 'Class'])

# Plot 1 Variance vs. Skewness - colored by Class
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Variance', y='Skewness', hue='Class', alpha=0.5)
plt.title('Variance vs. Skewness of Banknote')
plt.xlabel('Variance')
plt.ylabel('Skewness')
plt.legend(title='Class', loc='upper right')
plt.show()

# Plot 2 Curtosis vs. Entropy - colored by Class
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Curtosis', y='Entropy', hue='Class', alpha=0.6)
plt.title('Curtosis vs. Entropy of Banknote')
plt.xlabel('Curtosis')
plt.ylabel('Entropy')
plt.legend(title='Class', loc='upper right')
plt.show()

# Plot 3 Variance Distribution - Histogram by Class
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Variance', hue='Class', bins=50, kde=True) # Removed log_scale
plt.title('Variance Distribution by Class')
plt.xlabel('Variance')
plt.ylabel('Frequency')
plt.show()

# Plot 4Skewness Distribution by Class
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Skewness', hue='Class', bins=50, kde=True)
plt.title('Skewness Distribution by Class')
plt.xlabel('Skewness')
plt.ylabel('Frequency')
plt.show()

# Plot 5: Correlation Heatmap
plt.figure(figsize=(10, 8))  #fix figure size
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=True, linewidths=0.5) # annot=True
plt.title('Feature Correlation Heatmap')
plt.show()

# Plot 6: Boxplot of Variance by Class
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Class', y='Variance')
plt.title('Variance by Class')
plt.xlabel('Class (0 = Genuine, 1 = Forged)') #0 is genuine, 1 is forged
plt.ylabel('Variance')
plt.show()
```
