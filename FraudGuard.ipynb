```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Cole Manchester
# Load your dataset
df = pd.read_csv("creditcard.csv")

# Plot 1 Amount vs. Time (colored by Class)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Time', y='Amount', hue='Class', alpha=0.5)
plt.title('Transaction Amount vs. Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount ()')
plt.legend(title='Class', loc='upper right')
plt.show()

# Plot 2 V1 vs V2 (colored by Class)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='V1', y='V2', hue='Class', alpha=0.6)
plt.title('V1 vs V2 by Class')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend(title='Class', loc='upper right')
plt.show()

# Plot 3 Amount Distribution (Histogram plot thing)
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Amount', hue='Class', bins=50, kde=True, log_scale=(False, True))
plt.title('Transaction Amount Distribution by Class')
plt.xlabel('Amount ()')
plt.ylabel('Frequency')
plt.show()

# Plot 4 Time Distribution by Class
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Time', hue='Class', bins=100, kde=True)
plt.title('Transaction Time Distribution by Class')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency')
plt.show()

# Plot 5 Correlation Heatmap
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()

# Plot 6 Boxplot of Amount by Class
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Class', y='Amount')
plt.title('Transaction Amount by Class')
plt.xlabel('Class (0 = Fraud, 1 = Not Fraud)')
plt.ylabel('Amount ()')
plt.show()
```
